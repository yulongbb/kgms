```python
import urllib.request
import urllib.parse
from lxml import etree
import pandas as pd
from tqdm import tqdm

def handle(content, res_list):
    if len(res_list) == 0:
        return "Null"
    f = open("relation_1833.txt", "a+")
    count = 0
    for i in res_list:
        if i != "\n":
            if count % 2 == 0:
                relation = content+  "\t" + i
            else:
                pair = relation + "\t" + i + "\n"
                f.write(pair)
            count += 1
    f.close()
    return "Success"


def query(content):
    # 请求地址
    url = 'https://baike.baidu.com/item/' + urllib.parse.quote(content)
    # 请求头部
    headers = { 
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36' 
    }
    # 利用请求地址和请求头部构造请求对象
    req = urllib.request.Request(url=url, headers=headers, method='GET')
    # 发送请求，获得响应
    response = urllib.request.urlopen(req)
    # 读取响应，获得文本
    text = response.read().decode('utf-8')
    html = etree.HTML(text)
    # 构造 _Element 对象
    # 过滤数据，去掉空白
    sen_list = html.xpath('//a[contains(@class,"lemma-relation-link")]//text()')
    if len(sen_list) == 0:
        sen_list = html.xpath('//ul[contains(@class,"custom_dot para-list list-paddingleft-1")]//@href')
        if len(sen_list) == 0:
            return "No Multi"
        new_url = url+"/"+sen_list[0].split("/")[-1]
        req = urllib.request.Request(url=new_url, headers=headers, method='GET')
        response = urllib.request.urlopen(req)
        text = response.read().decode('utf-8')
        html = etree.HTML(text)
        sen_list = html.xpath('//a[contains(@class,"lemma-relation-link")]//text()')
    return handle(content, sen_list)

if __name__ == '__main__':
    data = pd.read_csv("raw_text.csv")
    gods = list(data['name'])
    for i, god in enumerate(gods):
        result = query(god)
        print("查询结果:",god, ":", result, i, "/", len(gods))
```
